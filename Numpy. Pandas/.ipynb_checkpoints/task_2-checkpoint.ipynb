### Задание 1

```python
import pandas as pd
```

```python
a = {"author_id": [1, 2, 3], "author_name": ['Тургенев', 'Чехов', 'Островский']}
a
```

```python
authors = pd.DataFrame(a)
authors
```

```python
b = {
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]
}
b
```

```python
books = pd.DataFrame(b)
books
```

### Задание 2

```python
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

### Задание 3

```python
top5 = authors_price.nlargest(5, 'price')
top5
```

### Задание 4

```python
group = authors_price.groupby('author_name')
group_min = group.agg({'price': "min"})
group_min = group_min.rename(columns={'price': 'min_price'})
group_min
```

```python
group_max = group.agg({'price': "max"})
group_max = group_max.rename(columns={'price': 'max_price'})
group_max
```

```python
group_mean = group.agg({'price': "mean"})
group_mean = group_mean.rename(columns={'price': 'mean_price'})
group_mean
```

```python
authors_stat = pd.concat([group_min, group_max, group_mean], axis=1)
authors_stat
```

```python
cover = pd.DataFrame({'cover': ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']})
cover
```

```python
authors_price = pd.concat([authors_price, cover], axis=1)
authors_price
```

```python
import numpy as np
book_info = pd.pivot_table(authors_price, values='price', index=['author_name'],
                          columns=['cover'], aggfunc=np.sum, fill_value=0)
book_info
```
